## Now we have the candidate .pkl file. Lets read it and cut a mini SDM from the big one, containing just the burst snippet, and dedisperse it 


```python
from rfpipe import candidates
import pylab as plt
import numpy as np
import glob
from loc_utils import get_maxsnr_cd, calc_cand_integration
%matplotlib inline
```

```python
# Reading the candidate .pkl file created in the refinement step 
pkls = glob.glob("*pkl")
pkls
pkl=pkls[0]
```

```python
# candcollection and #canddata

ccs = list(candidates.iter_cands(pkl, select='candcollection'))
cds = list(candidates.iter_cands(pkl, select='canddata'))
cds = cds[0]
cc = ccs[0]
cd = cds[np.argmax(cc.snrtot)]  # gets the data for the max SNR candidate. Here we only have one. So it's redundant

cd.snr1
```

```python
#Calculates the integration of the candidate using the canddata, after calculating for dm delay
#The integration is calcualted with respect to the start of the SDM, referenced to the max frequency of the band

i, name, dm = calc_cand_integration(cd)  # i is the cand_sample
print(i, name, dm)
```

```python
#Samples to cut the SDM. n is the number of integrations, i is the start sample(integartion)
n = 16
i = i - 8 # start sample = cand_sample - 10
i, n
```

```python
# Just creating a name for the DDCUT mini SDM
ext = f'ddcut{i}to{i+n-1}_{dm}'
ddcut_name = name[:-1] + '.' + f'{ext}'
print(ddcut_name)
```
